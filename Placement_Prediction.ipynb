Confusion Matrix

---
A confusion matrix is a simple and a useful tool for understading the performance of a classifiaction model,like one used in machine learning and statistics





WhatsApp Image 2025-09-12 at 7.21.45 PM.jpeg

00 → TN (True Negative)

01 → FP (False Positive)

10 → FN (False Negative)

11 → TP (True Positive)

1.Precision

Precision=TP/(TP+FP)

TP = True Positives

FP = False Positives

Measures: Out of all predicted positives, how many are actually positive?

2.Recall (Sensitivity, True Positive Rate)

Recall=TP/(TP+FN)

FN = False Negatives

Measures: Out of all actual positives, how many did we correctly predict?

3.F1 Score (harmonic mean of precision and recall)

F1_Score=2*Precision.Recall/(Precision+Recall)

```python
# Import all the necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
import plotly.express as px
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn import metrics
from datetime import datetime

from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
from sklearn.metrics import precision_recall_curve

import lightgbm as lgb
import xgboost as xgb
from xgboost import XGBClassifier
```

```python
df=pd.read_csv('placement.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
x=df.drop('placement',axis=1)
y=df['placement']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
confusion_matrix(y_test,lr.predict(x_test))
```

```python
sns.heatmap(confusion_matrix(y_test,lr.predict(x_test)),annot=True)
```

```python
precision_score(y_test,lr.predict(x_test))*100
```

```python
recall_score(y_test,lr.predict(x_test))*100
```

```python
f1_score(y_test,lr.predict(x_test))*100
```
